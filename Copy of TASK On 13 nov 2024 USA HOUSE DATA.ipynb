```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read dataset
df=pd.read_csv('/content/USA Housing Dataset.csv')
df.head()
```

```python
df.columns
```

```python
#removing unwanted columns
df.drop(['date','street', 'city',
       'statezip', 'country'],axis=1,inplace=True)
```

```python
df
```

```python
#check data have misssing entries on not
df.isnull().sum()
```

```python
df.columns
```

```python
#make price column as target viriable
x=df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement',
       'yr_built', 'yr_renovated']]

y=df['price']
```

```python
#divide data for traing and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# import linear regresssion model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#fit the model
lr.fit(x_train,y_train)
```

```python
#predict using trainng model
y_pred=lr.predict(x_test)
```

```python
#Calculate Mean Squared Error (MSE) and R-squared (RÂ²)
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
